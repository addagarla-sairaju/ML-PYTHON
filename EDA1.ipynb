##Exploratory Data Analysis

```python
# Import libraries for data manipulation and plotting
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset from CSV into a DataFrame
delivery_data=pd.read_csv('amazon_delivery.csv')

# (Optional) Convert to DataFrame, though pd.read_csv already does this
delivery_data=pd.DataFrame(delivery_data)
```

```python
# Display the first 5 rows of the dataset
delivery_data.head()
```

```python
# Display summary of the dataset: column types, non-null count, and memory usage
delivery_data.info()
```

```python
# Generate summary statistics for numerical columns
delivery_data.describe()
```

```python
# Display the last 5 rows of the dataset
delivery_data.tail()
```

```python
# Display the  2 rows of the dataset
delivery_data.sample(2)
```

```python
# Count plot for 'Weather' column
plt.figure(figsize=(8, 6))  # Set plot size
ax = sns.countplot(x='Weather', data=delivery_data, color='#232f3e')  # Create count plot

# Annotate counts above bars
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'),
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center',
                xytext=(0, 9), textcoords='offset points', fontsize=10, color='black')

# Add title and labels
plt.title('Distribution of Weather')
plt.xlabel('Weather')
plt.ylabel('Frequency')
plt.show()
```

```python
# Pair plot for selected columns
color = '#232f3e'
 # Set color
pair_plot = sns.pairplot(delivery_data[['Agent_Age', 'Agent_Rating', 'Store_Latitude', 'Drop_Longitude']], diag_kind='kde')

# Customize plot colors
for ax in pair_plot.axes.flatten():
    if ax:
        for artist in ax.lines + ax.collections:
           # Adjust line/scatter colors
            artist.set_color(color)
        for artist in ax.patches:
          # Adjust patch colors
            artist.set_edgecolor(color)
            artist.set_facecolor(color)

# Add title and show plot
plt.suptitle('Pair Plot of Selected Columns', y=1.02)
plt.show()
```

```python
# Check for missing values in each column and their counts
delivery_data.isna().sum()

# Set the size of the plot
plt.figure(figsize=(10, 6))

# Create a bar plot of the missing value counts
delivery_data.isna().sum().plot(kind='bar', color="#232f3e")

# Add a title to the plot
plt.title('Missing Values in Delivery Data')

# Label the x-axis
plt.xlabel('Columns')

# Label the y-axis
plt.ylabel('Number of Missing Values')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Adjust layout to prevent overlapping elements
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Check dataframe NULL values
delivery_data.isna().sum()
```

```python
# Summary for Categorical Columns
# Create a dictionary summarizing the counts of each category for categorical columns
categorical_summary = {
    col: delivery_data[col].value_counts()
    for col in delivery_data.select_dtypes(include=['object']).columns
}

# Print a header for the categorical summary
print("\nCategorical Summary:")

# Loop through each categorical column and its summary
for col, summary in categorical_summary.items():

    # Print the column name and its value counts
    print(f"\n{col.capitalize()}:\n", summary)
```

```python
# Get the order of weather categories based on their frequency (highest to lowest)
category_order = delivery_data['Weather'].value_counts().index

# Set the plot size
plt.figure(figsize=(10, 10))

# Create a horizontal bar plot for the 'Weather' column, ordered by frequency
ax = sns.countplot(data=delivery_data, y="Weather", order=category_order)

# Add counts as labels to each bar
for p in ax.patches:
    ax.annotate(f'{p.get_width()}', \
                ha='center', va='center',
                xytext=(25, 0), textcoords='offset points')

# Display the plot
plt.show()
```
