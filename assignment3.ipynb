##Basic Line Plot
*  Plot a simple line graph of ùë¶ = ùë•2 for ùë• values ranging from -10 to 10.
* Label the x-axis as "X Values" and the y-axis as "Squared Values."
* Add a title to the graph.


```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(-10, 11, 1)
y=x**2
plt.plot(x, y)
plt.xlabel("X Values")
plt.ylabel("Squared Values")
plt.title("Line Graph of y = x^2")


plt.show()
```

##Bar Plot
* Create a bar chart showing the marks scored by 5 students: [Sai, 'Vamsi',
'Divij', 'Sritan', 'Sita'] with marks [85, 90, 78, 92, 88].
* Add labels to the x-axis and y-axis and give the chart a title.

```python
import matplotlib.pyplot as plt
students=['Sai', 'Vamsi', 'Divij', 'Sritan', 'Sita']
marks=[85, 90, 78, 92, 88]
plt.bar(students, marks, color='skyblue')
plt.xlabel('Students')
plt.ylabel('Marks')
plt.title('Marks Scored by Students')
plt.show()
```

##Scatter Plot
* Generate two lists of 20 random integers each (for x and y coordinates).
* Create a scatter plot with these values, and customize it by adding color,
size, and labels to the points.

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.random.randint(1, 100, 20)
y = np.random.randint(1, 100, 20)
sizes = np.random.randint(50, 300, 20)
colors = np.random.rand(20)
plt.scatter(x, y, s=sizes, c=colors, alpha=0.6, cmap='viridis')
plt.xlabel('X Coordinates')
plt.ylabel('Y Coordinates')
plt.title('Customized Scatter Plot')
plt.show()
```

##Histogram
* Simulate 100 random values from a normal distribution using
np.random.randn() and plot a histogram.
* Customize the number of bins and add a grid to the plot.

```python
import matplotlib.pyplot as plt
import numpy as np
data = np.random.randn(100)
plt.hist(data, bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of 100 Random Values from Normal Distribution')
plt.grid(True)
plt.show()
```

##Pie Chart
* Visualize the percentage distribution of five categories ['A', 'B', 'C', 'D', 'E']
with values [20, 30, 25, 15, 10].
* Display the percentage values on the chart and use a color palette for
differentiation.

```python
import matplotlib.pyplot as plt
categories=['A','B','C','D','E']
values=[20,30,25,15,10]
plt.pie(values, labels=categories, autopct='%1.1f%%', colors=plt.cm.Paired.colors)
plt.title('Percentage Distribution of Categories')
plt.show()
```

##Box Plot (Seaborn)
* Create a box plot to visualize the distribution of the marks scored by
students in a class: [78, 85, 92, 88, 76, 95, 89, 84, 91, 73].

##Heatmap (Seaborn)
* Create a 5x5 matrix with random integers between 1 and 10 using
np.random.randint() and visualize it as a heatmap using Seaborn.
* Add a color bar and annotate the heatmap with the values.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
matrix=np.random.randint(1, 11, size=(5, 5))
sns.heatmap(matrix, annot=True, cmap='YlGnBu')
plt.title('Heatmap of Random 5x5 Matrix')
plt.show()
```

## Comparison of Matplotlib and Seaborn
* Generate a pair of plots (bar charts or line plots) using the same data in
both Matplotlib and Seaborn.
* Highlight the differences in the code, syntax, and aesthetics of the two
libraries.

```python
import matplotlib.pyplot as plt
import seaborn as sns
students=['Sai', 'Vamsi', 'Divij', 'Sritan', 'Sita']
marks=[85, 90, 78, 92, 88]
plt.bar(students, marks, color='blue')
plt.title("Matplotlib Bar Plot")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
sns.barplot(x=students, y=marks, palette='Blues')
plt.title("Seaborn Bar Plot")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

## Pair Plot (Seaborn)
* Manually create a small dataset with columns such as 'Age', 'Height',
'Weight' for 10 individuals and visualize the relationships using a pair plot.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data={
    'Age': [22, 25, 30, 35, 40, 45, 50, 55, 60, 65],
    'Height': [165, 170, 175, 180, 185, 190, 195, 200, 205, 210],
    'Weight': [55, 60, 65, 70, 75, 80, 85, 90, 95, 100]
}
df=pd.DataFrame(data)
sns.pairplot(df)
plt.title('Pair Plot of Age, Height, and Weight', y=1.02)
plt.show()
```
